part4-Train-test Split


```python
#Training data(80%)-> used to teach model
#Testing data(20%)->used to check how well the model performs on unseen data
from sklearn.model_selection import train_test_split
# Features
x = df[['Study_Hours', 'Attendance', 'Assignments']]
# Label
y = df['Marks']
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

Part5-Train Your first model


```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
print(model)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
print("Slope(coefficients):")
print(model.coef_)
```

```python
print("Intercept:")
print(model.intercept_)
```

part6-Predictions

```python
predictions=model.predict(x_test)
print(predictions)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x_train,y_train)
new_data=pd.DataFrame({
  'Study_Hours':[5,8,12],
  'Attendance':[80,90,95],
  'Assignments':[5,8,10]

})
model.fit(x_train,y_train)
predictions=model.predict(new_data)
print("Predicted Marks:")
print(predictions)
```

part7-Actual vs Predicted Comparison

```python
y_pred=model.predict(x_test)
import pandas as pd
comparison = pd.DataFrame({
    'Actual Marks': y_test,
    'Predicted Marks': y_pred,
    'Difference': abs(y_test - y_pred)
})

print(comparison)
Difference= y_test-y_pred
```

```python
print(comparison['Difference'].max())
```

part8-Model Evaluation


```python
from sklearn.metrics import mean_absolute_error, r2_score
y_pred=model.predict(x_test)
mae=mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error:",mae)
r2=r2_score(y_test,y_pred)
print("R^2 Score:",r2)
```
